```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
x=np.arange(11)
y1=[3651,5402,6119,10035,20882,17134,35412,470445,1175975,2658294,7868723]
y2=[3109,3063,2735,9426,17937,11744,27681,420351,859530,1952041,4944356]
plt.bar(x,y1,width=0.25,color='pink',label='IMPORT',log=True)
plt.bar(x+0.25,y2,width=0.25,color="k",label='EXPORT',log=True)
plt.legend(loc='upper left')
plt.xticks(x,['1951-56','1956-61','1961-66','1969-79','1980-85','1985-90','1992-97'
              ,'1997-2002','2002-07','2007-12'],rotation=45)
plt.xlabel('YEAR/PERIOD')
plt.ylabel('IMPORT & EXPORT (IN CRORE)')
plt.title('VOLUME OF INDIAN FOREIGN TRADE DURING DIFFERENT PERIOD')
plt.show()
```
